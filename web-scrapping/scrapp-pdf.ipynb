```python
#open the pdf file
file = open('/home/fatim/Documents/corpus/scrapping-text/1809070304_2016-01-10_BUKU_Pers,_Hukum,_&_Hak_Asasi_Manusia.pdf', 'rb')
reader = PyPDF2.PdfReader(file)
len(reader.pages)
```

```python
def text_split(input):
    all_text=[]
    for page in range(11,190):
        page = reader.pages[page]
        text = page.extract_text()
        all_text.append(text)

    text_join = ' '.join(all_text)
    split_text = text_join.split('\n')
    join_again = ' '.join(split_text)
    split_again = join_again.split('.')

    return split_again
```

```python
test = text_split(reader)
```

```python
import pandas as pd
test = pd.DataFrame({"transcript": test})
test
```

```python
import re

def filter1 (input):
    lower_text = []
    spesial_characters= r'[\=\-\/\:\;\(\) ]'
    quote = r'"'
    for i in input:
        j = i.lower()
        lower_text.append(j)

    all_text = []
    for i in lower_text:
        text = re.sub(r'\([^)]+\)', '', i)
        text = re.sub(spesial_characters,' ',i)
        text = re.sub(quote,' ', text)
        # text = re.sub(r'\b(?:[12][0-9]{2}|[1-9][0-9]?|300)\b', ' ', text)
        text = re.sub(r'[a-z]\.', '',text)
        text = re.sub(r'[A-Z]\.', '',text)
        text = re.sub(r'[1-9]\.', '',text)
        text = re.sub(r'[a-z]\)', '',text)
        text = re.sub(r'[A-Z]\)', '',text)
        text = re.sub(r'[1-9]\)', '',text)
        text = re.sub(r'“|”', ' ', text)
        text = re.sub(r"\'", ' ',text)
        all_text.append(text)
    return all_text
```

```python
test2 = filter1(test['transcript'])
test2
```
