```python
import requests
from bs4 import BeautifulSoup
```

```python
#scrapping liputan-6
all_text = []
for i in range(1,4):
  url = f"https://www.liputan6.com/bola/indeks?page={i}"
  response = requests.get(url)
  soup = BeautifulSoup(response.text, "html.parser")

  html = soup.find_all("article", "articles--rows--item")
  
  for i in html:
    text = i.find("div","articles--rows--item__summary").text
    all_text.append(text)
```

```python
#scrapping wikipedia korsel
url = "https://id.wikipedia.org/wiki/Korea_Selatan"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

html = soup.find_all("main", "mw-body")

new = []
for i in html:
  text = [j.get_text() for j in i.find_all("p")]
  new.append(text)
```
