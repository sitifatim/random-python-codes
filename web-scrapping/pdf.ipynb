```python
import PyPDF2
```

```python
file= open('BMPHukumAgraria.pdf', 'rb')
```

```python
reader = PyPDF2.PdfReader(file)
reader
```

```python
len(reader.pages)
```

```python
all_text=[]
for page in range(24,173):
    page = reader.pages[page]
    text = page.extract_text()
    all_text.append(text)
```

```python
all_text
```

```python
text_join = ''.join(all_text)
```

```python
split_text = text_join.split('\n')
```

```python
split_text
```

```python
join2 = ''.join(split_text)
join2
```

```python
split2 = join2.split('.')
split2
```

```python
fil_text = []
for i in range(len(split2)):
    t = split2[i].split()
    if len(t) > 9 and len(t) < 32:
        fil_text.append(split2[i])
```

```python
for i in fil_text:
    print(i)
    print('\n')
```

```python
filter_word = []
fil = ["segede","pertukaran"]
if fil in fil_text:
    filter_word.append(fil)
```
